# Define structure UnitCell

```python
#structure with 2 Li per formula unit
base_structure = ase.io.read('/nfshome/winkelmann/PycharmProjects/kmcCathodePython/program_files/00_LCO_R-3m_2fu_shaped_as_P21c.vasp')
end_repeat = [12, 8, 3]
```

### 25; 50 and 75%

```python
def write_structure_with_li_fraction(percentage, structure, repeat):
    structure = ase.build.sort(structure.repeat(repeat))
    structure.write(f'perfect_{percentage}pct_startstructure.XDATCAR', format='vasp-xdatcar', append=False, label='step %i time %.5e s' % (0, 0))
    
    # write out truncated file for WC analysis
    indices_truncated = [i for i, atom in enumerate(structure) if atom.symbol in ('Li', 'X')]
    new_atoms = structure[indices_truncated]
    new_atoms.write(f'perfect_{percentage}pct_startstructure_truncated.XDATCAR',
                    format='vasp-xdatcar', append=False)
```

```python
# 50%
structure_50pct = ase.build.sort(base_structure.copy())
li_indices = [i for i, atom in enumerate(structure_50pct) if atom.symbol == 'Li']
for i in range(1, len(li_indices), 2):
    structure_50pct[li_indices[i]].symbol = 'X'
write_structure_with_li_fraction(50, structure_50pct, end_repeat)
```

```python
#25%
repeat = (2, 2, 1)
end_repeat_25 = [round(x/y) for x, y in zip(end_repeat, repeat)]
structure_25pct = ase.build.sort(base_structure.copy().repeat(repeat))
li_indices = [i for i, atom in enumerate(structure_25pct) if atom.symbol == 'Li']
for i in range(0, len(li_indices)):
    if i%6 == 0:
        continue
    structure_25pct[li_indices[i]].symbol = 'X'
structure_25pct_full = write_structure_with_li_fraction(25, structure_25pct, end_repeat_25)
```

```python
#75%
repeat = (2, 2, 1)
end_repeat_75 = [round(x/y) for x, y in zip(end_repeat, repeat)]
structure_75pct = ase.build.sort(base_structure.copy().repeat(repeat))
li_indices = [i for i, atom in enumerate(structure_75pct) if atom.symbol == 'Li']
for i in range(0, len(li_indices)):
    if i%6== 0:
        structure_75pct[li_indices[i]].symbol = 'X'
structure_75pct_full = write_structure_with_li_fraction(75, structure_75pct, end_repeat_75)
```
