```python
import pandas as pd
import os
```

```python
#for two cutoff radii
best_value_file_name = '/nfshome/winkelmann/ARL/tmp/best_values_jochen.csv'
conv = 1000

#for files with diffrent lower cutoffs
cutoff_file = open('/nfshome/winkelmann/ARL/tmp/cutoffs_4_jochen_fit', 'r')  #edit for plotting over right cutoff
cutoff_file.readline()
lines = cutoff_file.readlines()
cutoff_file.close()
fit_methods = ['ardr-lambda', 'ardr-lineScan', 'rfe', 'lasso', 'least-squares', 'bayesian-ridge', 'elasticnet', 'omp',
               'ridge', 'split-bregman']
data = {}

cutoff_kinds = []
for line in lines:
    line = line.replace(' ', '').replace('\t','')
    line = line.split(',')
    if not line[1] in cutoff_kinds:
        cutoff_kinds.append(line[1])
    

for cutoff_kind in cutoff_kinds:
    data[cutoff_kind] = {}
    for fit_method in fit_methods:
        data[cutoff_kind][fit_method] = {'cutoff': [], 'validation': [], 'train': [], 'BIC': [], 'number': [], 'nonzero': []}

for line in lines:
    line = line.replace(' ', '').replace('\t','').replace('\n','')
    line = line.split(',')
    if len(line) > 9:
        print(line[8] + ',' + str(line[0:8]))
        continue
    data[line[1]][line[0]]['cutoff'].append(    float(line[2]))
    data[line[1]][line[0]]['validation'].append(float(line[3]))
    data[line[1]][line[0]]['train'].append(     float(line[4]))
    data[line[1]][line[0]]['BIC'].append(       float(line[5]))
    data[line[1]][line[0]]['number'].append(    float(line[6]))
    data[line[1]][line[0]]['nonzero'].append(   float(line[7]))

format = '%14s,\t%5s,\t%5s,\t%5s,\t%10s,\t%10s' + os.linesep
best_value_file = open(best_value_file_name, 'w')
best_value_file.write(format %('fit-algorithm', 'c2', 'c3', 'c4', 'RMSE_validation', 'RMSE_training'))
best_value_file.close()
best_values = {'fit_method': [], 'cutoff_kind': [], 'validation': []}

for fit_method in fit_methods:
    for cutoff_kind in cutoff_kinds:
        df = pd.DataFrame(data[cutoff_kind][fit_method])
        best_value_file = open(best_value_file_name, 'a')
        best_value_file.write(format %(fit_method, '8.09', cutoff_kind, df.cutoff[df.validation.idxmin()], df.validation.min()*conv, df.train[df.validation.idxmin()]*conv))
        best_value_file.close()
```
